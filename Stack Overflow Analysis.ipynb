# Data Cleaning

```python
#Converting Age and Years to numeric data types
#errors=coerse to deal with the string values in the data
public_analysis_df['Age1stCode']=pd.to_numeric(public_analysis_df.Age1stCode,errors='coerce')
public_analysis_df['YearsCode']=pd.to_numeric(public_analysis_df.YearsCode,errors='coerce')
public_analysis_df['YearsCodePro']=pd.to_numeric(public_analysis_df.YearsCodePro,errors='coerce')
```

```python
public_analysis_df.describe()
```

```python
#Age for programmers can't be 1 or 279.
public_analysis_df.drop(public_analysis_df[public_analysis_df.Age>100].index, inplace=True)
public_analysis_df.drop(public_analysis_df[public_analysis_df.Age<10].index, inplace=True)
```

```python
#Dropping columns where hours worked per week are >140
public_analysis_df.drop(public_analysis_df[public_analysis_df.WorkWeekHrs>140].index, inplace=True)
```

```python
#Keeping 3 gender options for simplicity(of the 6 given)
import numpy as np
public_analysis_df.where(~(public_analysis_df.Gender.str.contains(';',na=False)),np.nan,inplace=True)
```

```python
public_analysis_df['Gender'].value_counts()
```

# Analysis and Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
%matplotlib inline
```

```python
sns.set_style('darkgrid')
matplotlib.rcParams['figure.figsize']=(9,5)
matplotlib.rcParams['font.size']=14
matplotlib.rcParams['figure.facecolor']='#00000000'
```

```python
#Looking at the top 10 countries for this survey
Total_countries_involved=public_analysis_df.Country.nunique()
print('Total countries which participated in the survey:',Total_countries_involved)
```

```python
top_ten=public_analysis_df.Country.value_counts().head(10)
top_ten
```

```python
#Using a Bar chart to represent this result
plt.figure(figsize=(15,8))
plt.title('Top 10 countries in the survey')
sns.barplot(top_ten.index,top_ten);
```

```python
#Looking at the age group which participated in the survey
plt.figure(figsize=(15,8))
plt.title('Age-groups')
plt.xlabel('Age')
plt.ylabel('Number of respondents')
plt.hist(public_analysis_df.Age, bins=np.arange(10,100,5));
```

```python
#Gender visualization
gender_count=public_analysis_df.Gender.value_counts(dropna=False)
gender_count
```

```python
plt.figure(figsize=(15,8))
plt.title('Gender Chart')
plt.pie(gender_count,labels=gender_count.index,autopct='%1.1f%%',startangle=75);
```

```python
#Education Data
plt.figure(figsize=(12,12))
sns.countplot(y=public_analysis_df.EdLevel)
plt.title(schema_analysis_df['EdLevel'])
plt.ylabel(None);
```

```python
#Undergrad level %
undergrad_percent=public_analysis_df.UndergradMajor.value_counts()/public_analysis_df.UndergradMajor.count()*100
sns.barplot(undergrad_percent,undergrad_percent.index)
plt.title(schema_analysis_df['UndergradMajor'])
plt.ylabel(None);
plt.xlabel('Percentage');
```

```python
#Employment-type Data
(public_analysis_df.Employment.value_counts(normalize=True)*100).plot(kind='barh')
plt.title(schema_analysis_df.Employment)
plt.xlabel('Percentage')
plt.ylabel(None);
```
